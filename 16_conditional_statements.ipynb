```python
# in organisation you shouldnt write multiple if (rule)
## due to multiple ifs execution time inc and memory space taken is also much more
x=7
r=x%2
if r==1:
    print('Odd number')
else:
    print('even number')
```

```python
# if code has 1 condn-- if else only one time
# if code has more than 1 condn -- nested if
x=3
r=x%2
if r==1:
    print('Odd number')
    if x>5:
        print('greater number')
    else:
        print('smaller number')
else:
    print('even number')
```

```python
x=1
if x==1:
    print('one')
elif x==2:
    print('two')
elif x==3:
    print('three')
elif x==4:
    print('four')
else:
    print('number not found')
```
