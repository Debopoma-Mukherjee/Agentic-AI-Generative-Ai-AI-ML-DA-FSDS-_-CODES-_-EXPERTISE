```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
penguins["species"]
```

```python
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="island")
```

```python
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
#set_style() :The ways of styling themes are as follows:
#white
#dark
#whitegrid
#darkgrid
#ticks

#set_context():allows us to override default parameters.This affects things like the the labels.
#poster
#paper -ideal for printing in academic papers
#notebook -good for working inside jupyter notebooks(default setting).
#talk -good for presentation slides where readability matters
```

```python
sns.set_style("ticks")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_style("white")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_style("dark")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_context("poster")
sns.set_style("white")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_context("talk")
sns.set_style("white")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_context("notebook")
sns.set_style("white")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.set_context("paper")
sns.set_style("white")
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
#Swarm plot
sns.swarmplot(data=penguins,x="species",y="body_mass_g",hue="sex")  
```
