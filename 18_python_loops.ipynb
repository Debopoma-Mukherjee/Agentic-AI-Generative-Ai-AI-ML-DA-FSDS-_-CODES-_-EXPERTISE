```python
#We have another concept while else and for else
```

```python
i=5
while(i>=1):
    print("data science:",i)
    i=i-1
```

```python
i=1
while i<=5:
    print('data science')
    j=1
    while j<=4:
        print('technology')
        j=j+1
    i=i+1
    print()
        
```

```python
i=1
while i<=5:
    print('data science',end="")
    j=1
    while j<=4:
        print('technology',end="")
        j=j+1
    i=i+1
    print()
```

```python
i=1
while i<=2:
    j=0
    while j<=2:
        print(i*j,end=" ")
        j+=1
    print()
    i+=1
```

```python
i=1
while i<=4:
    j=0
    while j<=3:
        print(i*j,end=" ")
        j+=1
    print()
    i+=1
```

```python
i=10
while i<=4:
    j=0
    while j<=3:
        print(i*j,end=" ")
        j+=1
    print()
    i+=1
else:
    print('condition is not matched')
```

```python
#when to use while loop and when to use for loop
#while loop--use when we have condition
#for loop--use when we have a sequence--(datatype,datastructure)
```

```python
name="nit"
for i in name:
    print(i)
```

```python
for i in range(10,50,5):
    print(i)
```

```python
no=5
for i in range(1,11,1):
    print(i*no)    
```

```python
#break:if you apply break the loop will end
```

```python
for i in range(1,11):
    if i==6:
        break
    print(i)
```

```python
#continue
for i in range(1,11):
    if i==4:
        continue
    print(i)
```
