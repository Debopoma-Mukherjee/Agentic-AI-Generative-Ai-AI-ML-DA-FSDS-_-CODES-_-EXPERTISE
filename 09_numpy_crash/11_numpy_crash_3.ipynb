```python
import numpy as np
```

# 1.Array Creation Function

```python
# create an array from list
a=np.array([1,2,3])
print("Array a:\n",a)
```

```python
# create an array filled with zeros
d=np.zeros((2,3))
print("Array d:\n",d)
```

```python
# create an array fileed with ones
e=np.ones((3,2))
print("Array e\n",e)
```

```python
# create an identity matrix
f=np.eye(4)
print("Identity matrix f:\n",f)
```

# Array Manipulations

```python
#Reshape an array
a1=np.array([1,2,3])
reshape=np.reshape(a1,(1,3))
print("Reshaped array:",reshape)
```

```python
#Reshape an array
a1=np.array([1,2,3])
reshape=np.reshape(a1,(3,1))
print("Reshaped array:",reshape)
```

```python
#Flatten an array
f1=np.array([[1,2],[3,4]])
flattened=np.ravel(f1)
print("flattened array",flattened)
```

```python
#Transpose an array
e1=np.array([[1,2],[3,4]])
transposed=np.transpose(e1)
print("Transposed array\n",transposed)
```

```python
# Stack arrays vertically
a2=np.array([1,2])
b2=np.array([3,4])
stacked=np.vstack([a2,b2]) # stack a2 and b2 vertically
print("Stacked arrays:\n",stacked)
```
